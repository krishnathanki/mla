```python
"""
Plotting Fundamentals - Scatter Plots and Histograms Explored 
Scatter Plot is used when X -Axis Values effect on Y-Axis values need to be measured 
E.g. In this Money spent on Ads effect on Sales 
Histopgram is used to find the frequency or counbt of occurance of something 
E.g. How many times is the money spent on Ads between 50 to 100 million dollars ($) 
"""
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
data = pd.read_csv("/content/drive/MyDrive/_Teaching Material/Machine Learning /ML Material/ML_Jan to June 2022/ML Lab_Jan to June 2022/Lab #2_Linear Regression/Advertising.csv")

import matplotlib.pyplot as plt

plt.figure(figsize=(16, 10)) #Size of figure Length,Height. Change and see output 
plt.scatter(
 data['TV'],
 data['sales'],
 c='magenta' #change the color and see output 
)
plt.xlabel("Money spent on TV ads ($)")
plt.ylabel("Sales ($)")
plt.show()

```

```python
"""
Histopgram is used to find the frequency or counbt of occurance of something 
E.g. How many times is the money spent on Ads between 50 to 100 million dollars ($) 
"""
import matplotlib.pyplot as plt
#plt.figure(figsize=(16, 10)) #Size of figure Length,Height. Change and see output 
data.drop(['Unnamed: 0'], axis=1)
# Creating histogram
fig, ax = plt.subplots(figsize =(16, 6))
ax.hist(data['TV'], bins = [0, 25, 50, 75, 100, 125,150, 175, 200, 225, 250, 275, 300],align='left', color='orange', edgecolor='black',
              linewidth=2)
plt.xlabel("Money spent on TV ads ($)")
plt.ylabel("Count/Frequency")
# Show plot
plt.show()
```

```python
"""
Histopgram is used to find the frequency or counbt of occurance of something 
E.g. How many times is the money spent on Sales between 50 to 100 million dollars ($) 
"""

fig, ax = plt.subplots(figsize =(16, 6))
ax.hist(data['sales'], bins = [0, 5,10,15,20,25],align='left', color='yellow', edgecolor='black',linewidth=2)
plt.xlabel("Sales ($)")
plt.ylabel("Count/Frequency")
# Show plot
plt.show()
```
