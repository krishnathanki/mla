# **Basics Refresher - Use of Standard Scaler** 
When the range of values in your dataset is very high, larger values will overshadow the smaller values. 

To avoid this we use the **Standard Scaler** that will normalize all values between -1 to 1.

This part of data pre-processing. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.model_selection import train_test_split
data = pd.read_csv("/content/sample_data/Advertising.csv")
data.head(20)
```

```python
from sklearn.preprocessing import StandardScaler
data_copy = data.copy(deep = True)
data_copy = data_copy.drop(['sales', 'Unnamed: 0'], axis=1)
print(data_copy.shape)
sc_X = StandardScaler()
X =  pd.DataFrame(sc_X.fit_transform(data_copy),columns=['TV', 'radio', 'newspaper'])
print(X)
```
