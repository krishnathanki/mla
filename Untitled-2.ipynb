```python
import pandas
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
df=pandas.read_csv("Advertising.csv")
plt.scatter(df['sales'],df['TV'])
plt.show()
```

```python
from scipy import stats
slope,intercept,r,p,std_err=stats.linregress(df['sales'],df['TV'])
def myfunc(sales):
    return slope*sales+intercept
print("coefficient of corelation is:",r)
if r>=-0.5 and r<=0.5:
    print("Bad")
else:
    print("Good")
myModel= list(map(myfunc,df['sales']))
plt.scatter(df['sales'],df['TV'])
plt.plot(df['sales'],myModel)
plt.show()
```
